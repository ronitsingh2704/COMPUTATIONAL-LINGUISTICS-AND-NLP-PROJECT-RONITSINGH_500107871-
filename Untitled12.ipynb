```python
# Install required modules
import nltk
import string
import re
import pandas as pd
import matplotlib.pyplot as plt
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from wordcloud import WordCloud

# Download required NLTK data packages
nltk.download('punkt', force=True)
nltk.download('stopwords', force=True)
nltk.download('punkt_tab') # Download the punkt_tab data package

# Define the cleaning function
def cleanResume(resumeText):
    resumeText = re.sub('http\S+\s*', ' ', resumeText)  # remove URLs
    resumeText = re.sub('RT|cc', ' ', resumeText)  # remove RT and cc
    resumeText = re.sub('#\S+', '', resumeText)  # remove hashtags
    resumeText = re.sub('@\S+', '  ', resumeText)  # remove mentions
    resumeText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', resumeText)  # punctuation
    resumeText = re.sub(r'[^\x00-\x7f]', r' ', resumeText)  # non-ASCII
    resumeText = re.sub('\s+', ' ', resumeText)  # extra whitespace
    return resumeText

# Load the dataset
resumeDataSet = pd.read_csv('/content/UpdatedResumeDataSet.csv', encoding='utf-8')

# Stopwords
oneSetOfStopWords = set(stopwords.words('english') + ['``', "''"])
totalWords = []
Sentences = resumeDataSet['Resume'].values
cleanedSentences = ""

# Clean and tokenize first 160 resumes
for i in range(160):
    cleanedText = cleanResume(Sentences[i])
    cleanedSentences += cleanedText
    requiredWords = word_tokenize(cleanedText)
    for word in requiredWords:
        if word.lower() not in oneSetOfStopWords and word not in string.punctuation:
            totalWords.append(word.lower())

# Show most common words
wordfreqdist = nltk.FreqDist(totalWords)
mostcommon = wordfreqdist.most_common(50)
print("Most common words:\n", mostcommon)

# Generate and display word cloud
wc = WordCloud(width=800, height=800, background_color='white').generate(cleanedSentences)
plt.figure(figsize=(10, 10))
plt.imshow(wc, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
from sklearn.preprocessing import LabelEncoder

# Label encode the 'Category' column
le = LabelEncoder()
resumeDataSet['Category'] = le.fit_transform(resumeDataSet['Category'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from scipy.sparse import hstack  # Only needed if combining features later
import re

# Define the cleaning function
def cleanResume(resumeText):
    resumeText = re.sub('http\S+\s*', ' ', resumeText)  # remove URLs
    resumeText = re.sub('RT|cc', ' ', resumeText)  # remove RT and cc
    resumeText = re.sub('#\S+', '', resumeText)  # remove hashtags
    resumeText = re.sub('@\S+', '  ', resumeText)  # remove mentions
    resumeText = re.sub('[%s]' % re.escape("""!"#$%&'()*+,-./:;<=>?@[\]^_`{|}~"""), ' ', resumeText)  # punctuation
    resumeText = re.sub(r'[^\x00-\x7f]', r' ', resumeText)  # non-ASCII
    resumeText = re.sub('\s+', ' ', resumeText)  # extra whitespace
    return resumeText

# Apply the cleaning function to create the 'cleaned_resume' column if not exist
if 'cleaned_resume' not in resumeDataSet.columns:
    resumeDataSet['cleaned_resume'] = resumeDataSet.Resume.apply(lambda x: cleanResume(x))

# Feature and label setup
requiredText = resumeDataSet['cleaned_resume'].values
requiredTarget = resumeDataSet['Category'].values

# TF-IDF Vectorization
word_vectorizer = TfidfVectorizer(
    sublinear_tf=True,
    stop_words='english',
    max_features=1500
)
word_vectorizer.fit(requiredText)
WordFeatures = word_vectorizer.transform(requiredText)

print("TF-IDF feature extraction completed.")

# Splitting dataset into training and testing
X_train, X_test, y_train, y_test = train_test_split(
    WordFeatures, requiredTarget, random_state=0, test_size=0.2
)

print("Training shape:", X_train.shape)
print("Testing shape:", X_test.shape)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.multiclass import OneVsRestClassifier
from sklearn import metrics

# Train the classifier
clf = OneVsRestClassifier(KNeighborsClassifier())
clf.fit(X_train, y_train)

# Predict on test set
prediction = clf.predict(X_test)

# Accuracy on train and test
print('Accuracy of KNeighbors Classifier on training set: {:.2f}'.format(clf.score(X_train, y_train)))
print('Accuracy of KNeighbors Classifier on test set: {:.2f}'.format(clf.score(X_test, y_test)))

# Classification report
print("\nClassification report:\n")
print(metrics.classification_report(y_test, prediction))
```

```python
import pandas as pd
import re
import string
import nltk
from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
import ipywidgets as widgets
from IPython.display import display, clear_output

nltk.download('punkt')
nltk.download('stopwords')

# Clean resume/job description text
def clean_text(text):
    text = re.sub('http\S+\s*', ' ', text)
    text = re.sub('RT|cc', ' ', text)
    text = re.sub('#\S+|@\S+', '', text)
    text = re.sub('[%s]' % re.escape(string.punctuation), ' ', text)
    text = re.sub(r'[^\x00-\x7f]', r' ', text)
    text = re.sub('\s+', ' ', text)
    return text.strip().lower()

# Create input widgets
job_input = widgets.Textarea(
    value='',
    placeholder='Paste your job description here...',
    description='Job Desc:',
    layout=widgets.Layout(width='100%', height='120px')
)

num_resumes_input = widgets.BoundedIntText(
    value=2,
    min=1,
    max=10,
    step=1,
    description='Resumes:',
    disabled=False
)

resume_boxes = []
resume_inputs_container = widgets.VBox()

# Function to dynamically add text areas for resumes
def update_resume_inputs(change):
    resume_boxes.clear()
    resume_inputs_container.children = []
    for i in range(change['new']):
        name_input = widgets.Text(
            placeholder=f'Candidate {i+1} Name',
            description='Name:',
            layout=widgets.Layout(width='50%')
        )
        text_input = widgets.Textarea(
            placeholder=f'Paste resume text for Candidate {i+1}',
            layout=widgets.Layout(width='100%', height='100px')
        )
        resume_boxes.append((name_input, text_input))
    resume_inputs_container.children = [widgets.HBox([n, t]) for n, t in resume_boxes]

num_resumes_input.observe(update_resume_inputs, names='value')

# Score button
score_button = widgets.Button(description="Score Resumes", button_style='success')

# Scoring logic
def score_resumes(button):
    clear_output(wait=True)
    display(job_input, num_resumes_input, resume_inputs_container, score_button)

    job_description = clean_text(job_input.value)
    resumes = {}
    for name_widget, text_widget in resume_boxes:
        name = name_widget.value.strip() or "Unnamed"
        text = text_widget.value.strip()
        if text:
            resumes[name] = clean_text(text)

    if not job_description or not resumes:
        print("⚠️ Please provide both job description and at least one resume.")
        return

    resume_names = list(resumes.keys())
    resume_texts = list(resumes.values())

    vectorizer = TfidfVectorizer(stop_words='english')
    vectors = vectorizer.fit_transform([job_description] + resume_texts)
    similarities = cosine_similarity(vectors[0:1], vectors[1:]).flatten()

    results = pd.DataFrame({
        'Candidate': resume_names,
        'ATS Score (%)': [round(score * 100, 2) for score in similarities]
    }).sort_values(by='ATS Score (%)', ascending=False).reset_index(drop=True)

    best_match = results.iloc[0]['Candidate']

    print("=== ATS Scores ===")
    display(results)
    print(f"\n Best matching candidate: {best_match}")

score_button.on_click(score_resumes)

# Show the interactive form
display(job_input, num_resumes_input, resume_inputs_container, score_button)
update_resume_inputs({'new': num_resumes_input.value})
```
